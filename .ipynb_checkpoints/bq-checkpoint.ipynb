```python
import numpy as np
import GPy
import scipy.stats
import numpy as np
import scipy.integrate as integrate
def f_temp(x):
    return [scipy.stats.norm(3, 1).pdf(x[0])]
def f(x):
    return [scipy.stats.norm(3, 1).pdf(x[0]),scipy.stats.norm(0, 1).pdf(x[1])]
def flambda(a):
    return lambda x: np.cos(a*x)
def normal(s):
    scipy.stats.norm(0, 1).pdf(s)
```

```python
integrate.quad(flambda(0.0),np.NINF,np.inf)
```

```python
X = np.random.uniform(-10,10,size=(30,1))
Y = X.copy()
Y = np.apply_along_axis(flambda(1.0), 1, Y)
gp = GPy.models.GPRegression(X,Y,GPy.kern.src.rbf.RBF(input_dim=1))
```
